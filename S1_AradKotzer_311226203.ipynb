```python
# num 12
# 1) It can be seen that there is a relative correlation between the speed of travel (avg_speed) and the length of travel (detection_drives_count)
# Between the residents of Tel Aviv (0.875791) compared to a less strong correlation in Barcelona (0.214400)
#Maybe it could be because of more traffic jams in Barcelona
#It can be seen that there is a relatively high correlation between the number of turns made and the number of presses on the brakes.
#In Tel Aviv (0.754685), a stronger correlation can be seen than in Barcelona (0.594944)
#It may be that in Israel more people slow down before entering the round then in Barcelona
corr = Dmerge.corr(method ='pearson')
corr_Barca = detection_Barca.corr(method ='pearson',)
corr_Tel = detection_Tel_aviv.corr(method ='pearson')
corr3 =  pd.concat([corr_Barca, corr_Tel] )
corr3
```

```python
# create a new XLSX workbook
workbook = xlsxwriter.Workbook('311226203_Arad_kotzer1.xlsx')
worksheet = workbook.add_worksheet()

```

```python
#num 23

worksheet.write(0, 0, 'Index')
worksheet.write(0, 1, 'detection_drives_count')
worksheet.write(0, 2, 'avg_speed')
worksheet.write(0, 3, 'braking_count')
worksheet.write(0, 4, 'cornering_count')
```

```python
#num  24
# Loop through indices and rows in the dataframe using iterrows

for index,row in N_data1.iterrows():
    worksheet.write(index +1, 0, index + 1)
    worksheet.write(index +1, 1, N_data1.iloc[index][0])
    worksheet.write(index +1, 2, N_data1.iloc[index][1])
    worksheet.write(index +1, 3, N_data1.iloc[index][2])
    worksheet.write(index +1, 4, N_data1.iloc[index][3])
workbook.close()
```
